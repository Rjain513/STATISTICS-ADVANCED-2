#1 What is hypothesis testing in statistics?
Hypothesis testing is a statistical method used to determine whether there is enough evidence in a sample data to infer that a certain condition is true for the entire population. It involves formulating hypotheses and using sample data to decide which hypothesis is better supported.

#2 What is the null hypothesis, and how does it differ from the alternative hypothesis?
The null hypothesis (H 0 ) is a statement of no effect or no difference. The alternative hypothesis (H a or H 1) contradicts the null hypothesis, stating that there is a significant effect or difference. Hypothesis testing aims to determine if there's enough evidence to reject the H 0


#3 What is the significance level in hypothesis testing, and why is it important?
The significance level (α) is the probability of rejecting the null hypothesis when it is actually true (Type I error). It sets a threshold for determining statistical significance. Commonly used values are 0.05 and 0.01, indicating a 5% or 1% risk of a Type I error.

#4 What does a P-value represent in hypothesis testing?

#6 What are Type 1 and Type 2 errors in hypothesis testing?
A Type I error (false positive) occurs when we reject the null hypothesis when it is actually true. A Type II error (false negative) occurs when we fail to reject the null hypothesis when it is actually false. The probabilities of these errors are denoted by α and β, respectively.

#7 What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
A one-tailed test examines if the population parameter is greater than or less than a specific value, focusing on one direction of effect. A two-tailed test examines if the population parameter is different from a specific value, considering both directions of effect.

#8 What is the Z-test, and when is it used in hypothesis testing?
The Z-test is a statistical hypothesis test used to determine if there is a significant difference between the means of two groups or a sample mean and a population mean, when the population standard deviation is known and the sample size is large (typically n≥30).

#9 How do you calculate the Z-score, and what does it represent in hypothesis testing?
The Z-score is calculated as (x−μ)/σ, where x is a data point, μ is the population mean, and σ is the population standard deviation. It represents the number of standard deviations a data point is away from the mean in a standard normal distribution.

#10 What is the T-distribution, and when should it be used instead of the normal distribution?
The T-distribution is a probability distribution that is used when the population standard deviation is unknown and is estimated from the sample. It should be used instead of the normal distribution when dealing with small sample sizes (typically n<30) and unknown population variance.

#11 What is the difference between a Z-test and a T-test?
The key difference between a Z-test and a T-test lies in the knowledge of the population standard deviation. The Z-test assumes it's known, while the T-test is used when it's unknown and estimated from the sample. The T-test also accounts for the added uncertainty due to this estimation, resulting in a wider distribution for smaller sample sizes.

#12 What is the T-test, and how is it used in hypothesis testing?
The T-test is a statistical hypothesis test used to determine if there is a significant difference between the means of two groups or a sample mean and a population mean when the population standard deviation is unknown and estimated from the sample standard deviation. It is commonly used for smaller sample sizes.

#13 What is the relationship between Z-test and T-test in hypothesis testing?
Both Z-tests and T-tests are used to compare means in hypothesis testing. The Z-test is appropriate when the population standard deviation is known, while the T-test is used when it is unknown and estimated from the sample. As the sample size increases, the T-distribution approaches the standard normal distribution, making the results of both tests similar.

#14 What is a confidence interval, and how is it used to interpret statistical results?
A confidence interval is a range of values, derived from sample data, that is likely to contain the true value of a population parameter with a certain degree of confidence (e.g., 95%). It helps interpret statistical results by providing a plausible range for the parameter of interest.

#15 What is the margin of error, and how does it affect the confidence interval?
The margin of error is the range of values above and below the sample statistic within which the true population parameter is likely to fall. A larger margin of error results in a wider confidence interval, indicating less precision in the estimate. It is influenced by the sample size, variability, and confidence level.

#16 How is Bayes' Theorem used in statistics, and what is its significance?
Bayes' Theorem is a fundamental concept in probability theory that describes how to update the probability of a hypothesis based on new evidence. In statistics, it's used in Bayesian inference to revise prior beliefs in light of observed data, leading to posterior probabilities. Its significance lies in providing a framework for learning from data.

#18 What is the Chi-square goodness of fit test, and how is it applied?
The Chi-square goodness of fit test is used to determine if the observed frequencies of categorical data match an expected distribution. #19 What is the F-distribution, and when is it used in hypothesis testing?
The F-distribution is a continuous probability distribution that arises as the ratio of two independent chi-square random variables, each divided by its degrees of freedom. It is primarily used in hypothesis testing to compare variances of two or more populations, particularly in the context of ANOVA.

#20 What is an ANOVA test, and what are its assumptions?
An ANOVA (Analysis of Variance) test is a statistical test used to compare the means of two or more groups. Its key assumptions include independence of observations, normality of data within each group, and homogeneity of variances across the groups. Violations of these assumptions can affect the validity of the test results.

#21 What are the different types of ANOVA tests?
The different types of ANOVA tests include one-way ANOVA (comparing means of groups based on one independent variable), two-way ANOVA (examining the effects of two independent variables and their interaction on the means), and repeated measures ANOVA (used when the same subjects are measured multiple times).

#22 What is the F-test, and how does it relate to hypothesis testing?
The F-test is a statistical test that uses the F-distribution to compare the variances of two or more populations. In the context of hypothesis testing, particularly in ANOVA, the F-test is used to determine if there are any statistically significant differences between the means of the groups by examining the ratio of variances between and within the groups.

```python
#1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
import scipy.stats as stats

# Sample data
sample_mean = 10
sample_std = 2
sample_size = 30

# Population parameters
population_mean = 8

# Calculate the Z-statistic
z_statistic = (sample_mean - population_mean) / (sample_std / (sample_size**0.5))

# Calculate the p-value (two-tailed test)
p_value = 2 * stats.norm.sf(abs(z_statistic))

# Set the significance level (alpha)
alpha = 0.05

# Print the results
print("Z-statistic:", z_statistic)
print("P-value:", p_value)

# Interpret the results
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference between the sample mean and the population mean.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the sample mean and the population mean.")
```

```python
#3Implement a one-sample Z-test using Python to compare the sample mean with the population mean?

import math
import numpy as np
import scipy.stats as stats

def one_sample_z_test(sample_data, population_mean, population_std):


  sample_mean = np.mean(sample_data)
  sample_size = len(sample_data)
  z_statistic = (sample_mean - population_mean) / (population_std / math.sqrt(sample_size))
  p_value = 2 * stats.norm.sf(abs(z_statistic))  # Two-tailed test
  return z_statistic, p_value

# Example usage:
sample_data = [10, 12, 11, 13, 9, 10, 12, 11]
population_mean = 10
population_std = 2

z_statistic, p_value = one_sample_z_test(sample_data, population_mean, population_std)

print("Z-statistic:", z_statistic)
print("P-value:", p_value)

# Interpret the results (example with alpha = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference between the sample mean and the population mean.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the sample mean and the population mean.")
```

```python
#5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing?
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def visualize_errors(population_mean, population_std, sample_size, alpha, alternative_mean):

    # Calculate critical values for two-tailed test
    critical_value_lower = population_mean - stats.norm.ppf(1 - alpha / 2) * (population_std / np.sqrt(sample_size))
    critical_value_upper = population_mean + stats.norm.ppf(1 - alpha / 2) * (population_std / np.sqrt(sample_size))

    # Generate x-values for plotting
    x = np.linspace(population_mean - 4 * population_std, population_mean + 4 * population_std, 1000)

    # Calculate probability density functions
    pdf_null = stats.norm.pdf(x, population_mean, population_std / np.sqrt(sample_size))
    pdf_alternative = stats.norm.pdf(x, alternative_mean, population_std / np.sqrt(sample_size))

    # Plot distributions
    plt.plot(x, pdf_null, label='Null Hypothesis')
    plt.plot(x, pdf_alternative, label='Alternative Hypothesis')

    # Fill Type 1 error region
    plt.fill_between(x[x < critical_value_lower], 0, pdf_null[x < critical_value_lower], color='red', alpha=0.5, label='Type 1 Error')
    plt.fill_between(x[x > critical_value_upper], 0, pdf_null[x > critical_value_upper], color='red', alpha=0.5)

    # Fill Type 2 error region
    plt.fill_between(x[x > critical_value_lower], 0, pdf_alternative[x > critical_value_lower], color='blue', alpha=0.5)
    plt.fill_between(x[x < critical_value_upper], 0, pdf_alternative[x < critical_value_upper], color='blue', alpha=0.5, label='Type 2 Error')


    # Calculate Type 2 error probability
    beta = stats.norm.cdf(critical_value_upper, loc=alternative_mean, scale=population_std / np.sqrt(sample_size)) - \
           stats.norm.cdf(critical_value_lower, loc=alternative_mean, scale=population_std / np.sqrt(sample_size))

    # Add labels and title
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.title('Type 1 and Type 2 Errors')
    plt.legend()

    # Print Type 2 error probability
    print(f"Type 2 Error Probability (Beta): {beta:.4f}")

    # Show plot
    plt.show()

# Example usage:
visualize_errors(population_mean=100, population_std=15, sample_size=30, alpha=0.05, alternative_mean=110)
```

```python
#9 Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import scipy.stats as stats
import numpy as np

def calculate_confidence_interval(data, confidence_level=0.95):

    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # ddof=1 for unbiased sample standard deviation
    sample_size = len(data)

    # Calculate the critical value (t-score)
    critical_value = stats.t.ppf(1 - (1 - confidence_level) / 2, sample_size - 1)

    # Calculate the margin of error
    margin_of_error = critical_value * (sample_std / np.sqrt(sample_size))

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage:
data = [10, 12, 11, 13, 9, 10, 12, 11]
confidence_level = 0.95

lower_bound, upper_bound = calculate_confidence_interval(data, confidence_level)

print(f"Confidence Interval ({confidence_level*100:.0f}%): [{lower_bound:.2f}, {upper_bound:.2f}]")
```

```python
#11 Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
import numpy as np

def bayesian_inference(prior, likelihood, data):

    # Calculate the numerator of Bayes' Theorem (prior * likelihood)
    numerator = prior * likelihood(data)

    # Calculate the denominator of Bayes' Theorem (marginal likelihood)
    denominator = np.sum(numerator)

    # Calculate the posterior probability distribution
    posterior = numerator / denominator

    return posterior

# Example usage:
# Define the prior probability distribution (e.g., uniform distribution)
prior = np.array([0.5, 0.5])  # Two possible hypotheses

# Define the likelihood function (e.g., probability of data given hypothesis)
def likelihood(data):
    if data == 'heads':
        return np.array([0.8, 0.3])  # Likelihood for each hypothesis
    else:
        return np.array([0.2, 0.7])

# Observed data
data = 'heads'

# Perform Bayesian inference
posterior = bayesian_inference(prior, likelihood, data)

# Print the posterior probability distribution
print("Posterior Probability Distribution:", posterior)
```

```python
#13Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data
import numpy as np

def calculate_expected_frequencies(observed_frequencies):


  # Calculate row and column totals
  row_totals = observed_frequencies.sum(axis=1)
  col_totals = observed_frequencies.sum(axis=0)
  grand_total = observed_frequencies.sum()

  # Calculate expected frequencies
  expected_frequencies = np.outer(row_totals, col_totals) / grand_total

  return expected_frequencies

# Example usage:
observed_frequencies = np.array([[10, 20], [30, 40]])

expected_frequencies = calculate_expected_frequencies(observed_frequencies)

print("Observed Frequencies:")
print(observed_frequencies)
print("\nExpected Frequencies:")
print(expected_frequencies)
```

```python
#16 Implement an F-test using Python to compare the variances of two random samples?
import numpy as np
import scipy.stats as stats

# Generate two random samples (replace with your actual data)
np.random.seed(42)  # Set random seed for reproducibility
sample1 = np.random.normal(loc=10, scale=2, size=30)
sample2 = np.random.normal(loc=12, scale=3, size=40)

# Perform the F-test
f_statistic = np.var(sample1, ddof=1) / np.var(sample2, ddof=1)  # Calculate F-statistic
p_value = 1 - stats.f.cdf(f_statistic, len(sample1) - 1, len(sample2) - 1)  # Calculate p-value

# Print the results
print("F-statistic:", f_statistic)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis. The variances are not significantly different.")
```

```python
#17 Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results
import scipy.stats as stats
import numpy as np

# Sample data for three groups (replace with your actual data)
group1 = [10, 12, 11, 13, 9]
group2 = [15, 17, 16, 18, 14]
group3 = [20, 22, 21, 23, 19]

# Perform ANOVA test
fvalue, pvalue = stats.f_oneway(group1, group2, group3)

# Print the results
print("F-value:", fvalue)
print("P-value:", pvalue)

# Interpret the results
alpha = 0.05
if pvalue < alpha:
    print("Reject the null hypothesis. There is a significant difference between the means of the groups.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the means of the groups.")
```

```python
#19  Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVAD\
import scipy.stats as stats
import statsmodels.api as sm
import matplotlib.pyplot as plt

def check_anova_assumptions(data):

    # 1. Normality
    print("Normality Check (Shapiro-Wilk Test):")
    for i, group in enumerate(data):
        _, pvalue = stats.shapiro(group)
        print(f"Group {i+1}: p-value = {pvalue:.4f}")
        if pvalue < 0.05:
            print(f"  Warning: Normality assumption may be violated for Group {i+1}.")
        else:
            print(f"  Normality assumption is likely met for Group {i+1}.")

    # 2. Independence
    print("\nIndependence Assumption:")
    print("  Assumed to be met based on the experimental design.")

    # 3. Equal Variance (Homogeneity of Variance)
    print("\nEqual Variance Check (Levene's Test):")
    _, pvalue = stats.levene(*data)
    print(f"p-value = {pvalue:.4f}")
    if pvalue < 0.05:
        print("  Warning: Equal variance assumption may be violated.")
    else:
        print("  Equal variance assumption is likely met.")

    # Optional: Visualize normality using QQ-plots
    print("\nVisualizing Normality (QQ-plots):")
    fig, axes = plt.subplots(1, len(data), figsize=(12, 4))
    for i, group in enumerate(data):
        sm.qqplot(np.array(group), line='s', ax=axes[i])
        axes[i].set_title(f"Group {i+1}")
```

```python
#20  Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the RESULTS?
import statsmodels.formula.api as sm
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create sample data (replace with your actual data)
data = pd.DataFrame({'Factor1': ['A', 'A', 'B', 'B', 'A', 'A', 'B', 'B'],
                     'Factor2': ['X', 'Y', 'X', 'Y', 'X', 'Y', 'X', 'Y'],
                     'Value': [10, 12, 15, 18, 11, 13, 16, 19]})

# Perform two-way ANOVA
model = sm.ols('Value ~ Factor1 * Factor2', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Print ANOVA table
print(anova_table)

# Visualize the results
sns.pointplot(x='Factor1', y='Value', hue='Factor2', data=data)
plt.title('Interaction Effect of Factor1 and Factor2')
plt.show()
```

```python
#21Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing?
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Set parameters
degrees_of_freedom_numerator = [1, 3, 5, 10]  # Different degrees of freedom for numerator
degrees_of_freedom_denominator = [5, 10, 20]  # Different degrees of freedom for denominator
num_samples = 10000  # Number of samples to generate

# Generate random samples from F-distribution
for dfn in degrees_of_freedom_numerator:
    for dfd in degrees_of_freedom_denominator:
        samples = stats.f.rvs(dfn, dfd, size=num_samples)  # Generate samples

        # Plot the distribution
        plt.hist(samples, bins=50, density=True, alpha=0.5, label=f'dfn = {dfn}, dfd = {dfd}')

# Add theoretical PDF for comparison
x = np.linspace(0, 5, 100)
for dfn in degrees_of_freedom_numerator:
    for dfd in degrees_of_freedom_denominator:
        pdf = stats.f.pdf(x, dfn, dfd)
        plt.plot(x, pdf, label=f'Theoretical (dfn = {dfn}, dfd = {dfd})')

# Customize the plot
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('F-distribution')
plt.legend()
plt.show()
```

```python
#23 Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means?
import numpy as np
import scipy.stats as stats

# Set parameters for data simulation
population_mean = 100  # True population mean
population_std = 15   # True population standard deviation
sample_size = 30     # Sample size

# Simulate random data
np.random.seed(42)  # Set seed for reproducibility
data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Perform one-sample t-test
t_statistic, p_value = stats.ttest_1samp(data, popmean=population_mean)

# Print the results
print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Interpret the results (example with alpha = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference between the sample mean and the population mean.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the sample mean and the population mean.")
```

```python
#24  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
import numpy as np
import scipy.stats as stats

# Sample data (replace with your actual data)
data = [10, 12, 11, 13, 9, 10, 12, 11]

# Population variance (null hypothesis)
population_variance = 4

# Significance level
alpha = 0.05

# Calculate sample statistics
sample_size = len(data)
sample_variance = np.var(data, ddof=1)  # ddof=1 for unbiased sample variance

# Calculate the chi-square statistic
chi_square_statistic = (sample_size - 1) * sample_variance / population_variance

# Calculate the p-value (two-tailed test)
p_value = 2 * min(stats.chi2.cdf(chi_square_statistic, df=sample_size - 1),
                  1 - stats.chi2.cdf(chi_square_statistic, df=sample_size - 1))

# Print the results
print("Chi-square statistic:", chi_square_statistic)
print("P-value:", p_value)

# Interpret the results
if p_value < alpha:
    print("Reject the null hypothesis. The population variance is significantly different from", population_variance)
else:
    print("Fail to reject the null hypothesis. There is no significant evidence to suggest that the population variance is different from", population_variance)
```

```python
#25 Write a Python script to perform a Z-test for comparing proportions between two datasets or groups?
import numpy as np
import scipy.stats as stats

def z_test_proportions(successes1, trials1, successes2, trials2):


  # Calculate proportions
  p1 = successes1 / trials1
  p2 = successes2 / trials2

  # Calculate pooled proportion
  p_pooled = (successes1 + successes2) / (trials1 + trials2)

  # Calculate standard error
  se = np.sqrt(p_pooled * (1 - p_pooled) * (1 / trials1 + 1 / trials2))

  # Calculate Z-statistic
  z_statistic = (p1 - p2) / se

  # Calculate p-value (two-tailed test)
  p_value = 2 * stats.norm.sf(abs(z_statistic))

  return z_statistic, p_value

# Example usage:
successes1 = 120  # Number of successes in group 1
trials1 = 200      # Total trials in group 1
successes2 = 150  # Number of successes in group 2
trials2 = 300      # Total trials in group 2

z_statistic, p_value = z_test_proportions(successes1, trials1, successes2, trials2)

print("Z-statistic:", z_statistic)
print("P-value:", p_value)

# Interpret the results (example with alpha = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference between the proportions.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the proportions.")
```

```python
#26 Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results?
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generate two random samples (replace with your actual data)
np.random.seed(42)  # Set random seed for reproducibility
data1 = np.random.normal(loc=10, scale=2, size=30)
data2 = np.random.normal(loc=12, scale=3, size=40)

# Perform F-test
fvalue, pvalue = stats.f_oneway(data1, data2)

# Print the results
print("F-value:", fvalue)
print("P-value:", pvalue)

# Interpretation
alpha = 0.05  # Significance level
if pvalue < alpha:
    print("Reject the null hypothesis. The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis. The variances are not significantly different.")

# Visualization
plt.boxplot([data1, data2], labels=['Dataset 1', 'Dataset 2'])
plt.title('Comparison of Variances (F-test)')
plt.ylabel('Values')
plt.show()
```

```python
#27 Perform a Chi-square test for goodness of fit with simulated data and analyze the results.
import numpy as np
import scipy.stats as stats

# Generate simulated data (e.g., from a Poisson distribution)
np.random.seed(42)  # For reproducibility
observed_data = np.random.poisson(lam=5, size=100)  # Simulate 100 data points

# Define the expected distribution (e.g., Poisson with lambda=5)
expected_distribution = stats.poisson(mu=5)

# Calculate expected frequencies
unique_values, counts = np.unique(observed_data, return_counts=True)
expected_frequencies = expected_distribution.pmf(unique_values) * len(observed_data)

# Perform Chi-square goodness of fit test
chi_square_statistic, p_value = stats.chisquare(counts, f_exp=expected_frequencies)

# Print the results
print("Chi-square statistic:", chi_square_statistic)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis. The observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis. The observed data fits the expected distribution.")
```
